```python
import json
import csv
import pandas as pd
from tabulate import tabulate

json_path = 'C:/Users/urban/OneDrive/Documentos/GitHub/MiVOLO/jsons/test_detections.json'
csv_path = 'output.csv'
```

```python
with open(json_path, 'r') as json_file:
    data = json.load(json_file)

with open(csv_path, mode='w', newline='') as csv_file:
    fieldnames = ['uniqueId', 'classId', 'frame_number', 'bbox_x', 'bbox_y', 'bbox_width', 'bbox_height']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    
    writer.writeheader()

    for entry in data:
        # valores da bbox
        bbox_x, bbox_y, bbox_width, bbox_height = entry['bbox']
        
        # escrever csv
        writer.writerow({
            'uniqueId': entry['uniqueId'],
            'classId': entry['classId'],
            'frame_number': entry['frame_number'],
            'bbox_x': bbox_x,
            'bbox_y': bbox_y,
            'bbox_width': bbox_width,
            'bbox_height': bbox_height
        })

print(f'salvo com sucesso {csv_path}')
```

```python
df = pd.read_csv(csv_path)
print(tabulate(df, headers='keys', tablefmt='psql'))
```

```python
df_unique = df.drop_duplicates(subset=['uniqueId'])
# cont
num_unique_ids = df_unique['uniqueId'].nunique()

print(tabulate(df_unique, headers='keys', tablefmt='psql'))
print(f"\nTotal de pessoas nicas (IDs nicos): {num_unique_ids}")
```
